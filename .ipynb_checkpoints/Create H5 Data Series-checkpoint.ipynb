### Read CSV

```python
# Read original CSV
data_df = pd.read_csv("utils/datasets/standard_datasets/nyse_o.csv")
data_df.head()
```

```python
asset_list = data_df.columns.tolist()
```

```python
# For each asset
result = []
for a in asset_list:
    result_asset = []
    _open = 0
    for close in data_df[a]:
        if _open != 0:
            result_asset.append([_open, close, close, close, close])
        _open = close
    result.append(result_asset)        
```

```python
result = np.array(result)
```

### Save H5

```python
abb = [abbr.encode() for abbr in asset_list]
```

```python
write_to_h5py(result, abb, "utils/datasets/tse.h5")
```
