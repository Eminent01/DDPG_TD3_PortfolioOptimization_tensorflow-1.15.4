```python
# IMPORTS
import os
import numpy as np
from utils.data import read_stock_history, normalize
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import seaborn as sns
from model.non_ai import tools, algos
from model.non_ai.algos import *
from model.non_ai.result import AlgoResult, ListResult
from environment.portfolio import PortfolioEnv, MultiActionPortfolioEnv, create_close_dataframe, convert_prices
from stock_trading import test_portfolio_selection, test_portfolio_selection_multiple, plot_weights, plot_portfolio_values, results_table
from technical_indicators.technical_indicators import moving_average, exponential_moving_average, momentum, rate_of_change, bollinger_bands, trix, standard_deviation, full_rmr_moving_average, olmar_moving_average

print("Imports Complete!")
```

```python
# Dataset
history, abbreviation, date_list = read_stock_history(filepath='utils/datasets/nyse_n.h5')
print("Dataset Shape: {}".format(history.shape))
print("Dataset Dates: {} -> {}".format(date_list[0], date_list[-1]))
```

```python
# Get Close
_history = history[:, :, :4]
# Target Stocks
target_stocks = abbreviation
# Target History
target_history = np.empty(shape=(len(target_stocks), _history.shape[1], _history.shape[2]))
for i, stock in enumerate(target_stocks):
    target_history[i] = _history[abbreviation.index(stock), :, :]

print("Assets: ", target_stocks)
print("Target History Sample: ", target_history[0][0])
print("Target History Shape: ", target_history.shape)
```

```python
window_length = 3
```

```python
# Training/Testing Date Range
full_length = len(date_list)
train_test_ratio = 6/7
test_start_date = date_list[window_length]
test_end_date = date_list[full_length-2]
print("Testing Date Range: {} -> {} ({} Steps)".format(test_start_date, test_end_date, 
                                                    (int)(date_list.index(test_end_date) - date_list.index(test_start_date))))
```

```python
# Set Up Environment
env = PortfolioEnv(target_history, target_stocks, date_list, trading_cost=0, start_date=test_start_date, 
                              end_date=test_end_date, window_length=window_length, olps=False)
```

```python
# Plot
plt.figure(figsize=(10, 6), dpi=100)
plt.title('Portfolio Values')
plt.ylabel('Portfolio Value')
plt.xlabel('Day')
w = 3

plt.plot(env.close_df['JOHNSON&JOHNSON'], label='JOHNSON&JOHNSON')
#plt.plot(env.close_df['JOHNSON&JOHNSON'] / env.close_df['JOHNSON&JOHNSON'].shift(1) - 1, label='Return)
plt.plot(moving_average(env.close_df['JOHNSON&JOHNSON'], w), label='MA-{}'.format(w))
plt.plot(exponential_moving_average(env.close_df['JOHNSON&JOHNSON'], w), label='EMA-{}'.format(w))
#plt.plot(momentum(env.close_df['JOHNSON&JOHNSON'], w), label='Momentum-{}'.format(w))
#plt.plot(rate_of_change(env.close_df['JOHNSON&JOHNSON'], w), label='RoC-{}'.format(w)) 
Upper, Lower = bollinger_bands(env.close_df['JOHNSON&JOHNSON'], w)
plt.plot(Upper, label='Bollenger_upper-{}'.format(w)) 
plt.plot(Lower, label='Bollenger_lower-{}'.format(w)) 
plt.plot(standard_deviation(env.close_df['JOHNSON&JOHNSON'], 30), label='STD-{}'.format(w))

plt.xticks(np.arange(0, len(env.close_df), 100))
plt.xticks(rotation=90)
plt.legend(loc='upper center', bbox_to_anchor=(1.25, 1), shadow=True, ncol=1)
plt.show()
```

```python
# Close Price DataFrame for on-line Portfolio Selection
temp_close_df = create_close_dataframe(history, abbreviation, date_list)
# Add empty asset in the beginning
#temp_close_df.insert(0, 'Unassigned', 1)
temp_close_df = convert_prices(temp_close_df, 'raw', True)    
print('temp_close_df.shape: {}'.format(temp_close_df.shape))
temp_close_df.head()
```

```python
# Plot RMR and OLMAR Predictions
plt.figure(figsize=(10, 6), dpi=100)
plt.title('Portfolio Values')
plt.ylabel('Portfolio Value')
plt.xlabel('Day')
w = 3

full_rmr_moving_average_df = full_rmr_moving_average(temp_close_df, w)

plt.plot(temp_close_df['JOHNSON&JOHNSON'], label='JOHNSON&JOHNSON')
plt.plot(full_rmr_moving_average_df['JOHNSON&JOHNSON'], label='RMR-MA-{}'.format(w))
plt.plot(olmar_moving_average(temp_close_df['JOHNSON&JOHNSON'], w), label='OLMAR-MA-{}'.format(w))

plt.xticks(np.arange(0, len(env.close_df), 100))
plt.xticks(rotation=90)
plt.legend(loc='upper center', bbox_to_anchor=(1.25, 1), shadow=True, ncol=1)
plt.show()
```

```python
rmr_moving_average_df = full_rmr_moving_average(temp_close_df, w)
print(rmr_moving_average_df.head())
for i in range(len(abbreviation)):
    # Create List
    ti = []
    ti.append([olmar_moving_average(temp_close_df[abbreviation[i]], w)]) # Based on OLMAR
    ti.append([rmr_moving_average_df[abbreviation[i]]]) # Based on RMR

    # Turn to Array
    ti = np.vstack(ti)

    ti_reshaped = []
    for j in range(temp_close_df.shape[1]):
        ti_reshaped.append(ti[:, j])   

    technical_indicator_history.append(ti_reshaped)

technical_indicator_history = np.array(technical_indicator_history)    
print('technical_indicators.shape: {}'.format(technical_indicator_history.shape)) 
```
